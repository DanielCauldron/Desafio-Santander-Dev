.Extração (Extract) :

```python
import pandas as pd

# Caminho para o arquivo CSV
file_path = "Netflix Userbase.csv"
# Carregar o CSV em um DataFrame do Pandas
data = pd.read_csv(file_path)
```

```python
# visualizando
data.head(5)
```

. Transformação (Transform):

```python
# Exemplo de limpeza e transformação simples
data.drop_duplicates(inplace=True) #descarta linhas duplicadas 

data.dropna(inplace=True) #esta linha descarta linhas com valores ausentes

data['Date'] = pd.to_datetime(data['Date']) #  esta linha converte a coluna 'Date' no DataFrame para o formato datetime usando a pd.to_datetime()

data['Year'] = data['Date'].dt.year#esta linha cria uma nova coluna chamada 'Ano' no DataFrame extraindo o componente de ano da coluna 'Data' usando o .dt.year
```

```python
data.head()
```

. Carga (Load) :

```python
# Exemplo de carga para um arquivo CSV modificado
data.to_csv("Netflix_Userbase_Cleaned.csv", index=False)
```
